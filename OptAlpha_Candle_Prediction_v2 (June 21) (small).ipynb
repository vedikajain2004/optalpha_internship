```python
#clustering df
cdf = pd.DataFrame()
epsilon = 1e-10
bodylength = df['oc'] / cd['open']
cdf['candlelength'] = df['hl'] / cd['low']
cdf['bodypct'] = bodylength / (cdf['candlelength'] + epsilon)
lowerwick = (cd[['open', 'close']].min(axis=1) - cd['low']) / cd['open']
upperwick = (cd['high'] - cd[['open', 'close']].max(axis=1)) / cd['open']
cdf['uwpct'] = upperwick / (cdf['candlelength'] + epsilon)
cdf['lwpct'] = abs(lowerwick) / (cdf['candlelength'] + epsilon)
cdf['vol_1st_grad'] = df['vol_1st_grad']
#ccdf = cdf.dropna()
cdf.head(15)
```

```python
cdf.info()
```

```python
kmeans = KMeans(n_clusters=6, random_state=101)
kmeans.fit(cdf)
cdf['labels'] = kmeans.labels_

#this plot gives no information
'''u_labels = np.unique(labels)
pcdf = PCA(2).fit_transform(cdf)
for i in u_labels:
    plt.scatter(pcdf[labels == i, 0], pcdf[labels == i, 1], label = i)
plt.legend()
plt.show()'''

#visualisation is very important for clarity
cdf['index'] = cdf.index
'''cols = ['open', 'high', 'low', 'close']
temp_df = cd[cols]
temp_df['datetime'] = df['datetime']'''
temp_df = pd.DataFrame({'datetime': df['datetime'], 'open': df['opnorm'], 'high': df['hinorm'], 'low': df['lonorm'], 'close': df['clnorm']})
#non normalised prices didn't show any trends
temp_df['index'] = temp_df.index
vcdf = pd.merge(cdf, temp_df, on='index')
cols_ohlc = ['datetime', 'open', 'high', 'low', 'close', 'labels', 'vol_1st_grad']
temp_df = vcdf[cols_ohlc]
```
